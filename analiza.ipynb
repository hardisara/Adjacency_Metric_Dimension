```python
import csv
import json

def grafi_iz_csv(datoteka):
    with open(datoteka) as f:
        rd = csv.reader(f)
        for n, g6, dim, S in rd:
            yield (Integer(n), Graph(g6), Integer(dim), json.loads(S))
```

```python
grafi = list(grafi_iz_csv('grafi3-9.csv'))
drevesa = list(grafi_iz_csv('drevesa3-19.csv'))
```

```python
def izpisi(seznam, v=None, d=None):
    for n, G, dim, S in seznam:
        if (v is None or n == v) and (d is None or dim == d):
            G.show()
            print(S)
```

```python
# Primer: grafi na 5 vozliih dimenzije 2
izpisi(grafi, 5, 2)
```

```python
# Primer: drevesa na 10 vozliih dimenzije 8
izpisi(drevesa, 10, 8)
```
